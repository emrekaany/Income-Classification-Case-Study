```python
import numpy as np
import pandas as pd

df_model = df.copy()

# Strip spaces from string cols
str_cols = df_model.select_dtypes(include="object").columns
df_model[str_cols] = df_model[str_cols].apply(lambda s: s.str.strip())

# ----- target -----
df_model = df_model[~df_model["income"].isna()].copy()
df_model["income_high"] = (df_model["income"] == ">50K").astype(int)

# Drop leakage
df_model = df_model.drop(columns=["Name"])

# ----- age features -----
age_bins   = [0, 25, 35, 45, 55, 65, 120]
age_labels = ["<25", "25-34", "35-44", "45-54", "55-64", "65+"]
df_model["age_group"] = pd.cut(df_model["age"], bins=age_bins, labels=age_labels, right=False)

# ----- hours features -----
hours_bins   = [0, 35, 41, 46, 61, 200]
hours_labels = ["<35", "35-40", "41-45", "46-60", "60+"]
df_model["hours_group"] = pd.cut(df_model["hours-per-week"], bins=hours_bins,
                                 labels=hours_labels, right=False)
df_model["long_hours_flag"] = (df_model["hours-per-week"] >= 46).astype(int)

# ----- education features -----
edu_order = [
    "Preschool","1st-4th","5th-6th","7th-8th","9th","10th","11th","12th",
    "HS-grad","Some-college","Assoc-voc","Assoc-acdm",
    "Bachelors","Masters","Prof-school","Doctorate"
]
edu_map = {edu: i for i, edu in enumerate(edu_order)}
df_model["education_level"] = df_model["education_level"].fillna("Unknown")
df_model["education_score"] = df_model["education_level"].map(edu_map).fillna(-1)

high_edu_levels = {"Bachelors","Masters","Prof-school","Doctorate"}
df_model["high_education_flag"] = df_model["education_level"].isin(high_edu_levels).astype(int)

# ----- capital features -----
df_model["has_capital_gain"] = (df_model["capital-gain"] > 0).astype(int)
df_model["has_capital_loss"] = (df_model["capital-loss"] > 0).astype(int)
df_model["log_capital_gain"] = np.log1p(df_model["capital-gain"])
df_model["log_capital_loss"] = np.log1p(df_model["capital-loss"])

# ----- occupation / workclass -----
high_occ = {"Exec-managerial", "Prof-specialty"}
df_model["is_exec_prof"] = df_model["occupation"].isin(high_occ).astype(int)

def group_workclass(w):
    if w in {"State-gov","Local-gov","Federal-gov"}:
        return "Gov"
    if w == "Private":
        return "Private"
    if w in {"Self-emp-inc", "Self-emp-not-inc"}:
        return w
    return "Other"

df_model["workclass_group"] = df_model["workclass"].map(group_workclass)

# ----- marital / relationship -----
married_statuses = {"Married-civ-spouse", "Married-AF-spouse", "Married-spouse-absent"}
df_model["is_married"]   = df_model["marital-status"].isin(married_statuses).astype(int)
df_model["is_in_couple"] = df_model["relationship"].isin({"Husband","Wife"}).astype(int)
df_model["has_children"] = df_model["relationship"].isin({"Own-child","Other-relative"}).astype(int)

# ----- geography -----
df_model["is_US"] = (df_model["native-country"] == "United-States").astype(int)

# Simple region grouping example
europe = {"England","Germany","Portugal","France","Italy","Ireland","Scotland","Greece","Holand-Netherlands"}
latin_america = {"Mexico","Cuba","Jamaica","Honduras","Puerto-Rico","Columbia","Ecuador",
                 "Trinadad&Tobago","Peru","Nicaragua","El-Salvador","Guatemala","Haiti","Dominican-Republic"}
asia = {"Philippines","India","Japan","China","Vietnam","Laos","Cambodia","Thailand","Taiwan","Hong"}

def map_region(c):
    if c in europe: return "Europe"
    if c in latin_america: return "LatAm"
    if c in asia: return "Asia"
    if c == "United-States": return "US"
    return "Other"

df_model["country_region"] = df_model["native-country"].map(map_region)

# ----- interactions -----
df_model["high_edu_mid_age"] = (
    (df_model["high_education_flag"] == 1) &
    df_model["age"].between(35, 54)
).astype(int)

df_model["exec_prof_long_hours"] = (
    (df_model["is_exec_prof"] == 1) &
    (df_model["long_hours_flag"] == 1)
).astype(int)

df_model["investor_flag"] = (
    (df_model["has_capital_gain"] == 1) |
    (df_model["has_capital_loss"] == 1)
).astype(int)
```

```python
cat_cols = [
    "workclass", "education_level", "marital-status",
    "occupation", "relationship", "race", "sex", "native-country"
]

def show_cat_target_stats(df, col, target="income_high", min_count=100):
    tab = (
        df.groupby(col)[target]
          .agg(["mean", "count"])
          .rename(columns={"mean": "income_high_rate", "count": "n"})
          .sort_values("income_high_rate", ascending=False)
    )
    print(f"\n=== {col}: income_high rate (min_count = {min_count}) ===")
    print(tab[tab["n"] >= min_count])

for col in cat_cols:
    print(f"\n=== {col} value counts (known incomes) ===")
    print(df_known[col].value_counts().head(10))  # top categories
    show_cat_target_stats(df_known, col)
```

```python
def rate_by_group(df, group_col, target="income_high", min_count=100):
    tab = (
        df.groupby(group_col)[target]
          .agg(["mean", "count"])
          .rename(columns={"mean": "income_high_rate", "count": "n"})
          .sort_values("income_high_rate", ascending=False)
    )
    return tab[tab["n"] >= min_count]

print("\n=== income_high rate by sex ===")
print(rate_by_group(df_known, "sex"))

print("\n=== income_high rate by race ===")
print(rate_by_group(df_known, "race"))
```

```python
prev_df = pd.DataFrame({
    "dataset": [
        "Labeled sample", "Labeled sample",
        "Bank population\n(expected)", "Bank population\n(expected)"
    ],
    "income_group": [">50K", "<=50K", ">50K", "<=50K"],
    "share": [
        pi_sample,
        1 - pi_sample,
        pi_bank,
        1 - pi_bank
    ]
})

fig, ax = plt.subplots(figsize=(7, 4))
sns.barplot(
    data=prev_df,
    x="dataset",
    y="share",
    hue="income_group",
    ax=ax
)
ax.set_ylabel("Share of customers")
ax.set_xlabel("")
ax.set_title("Income distribution: training sample vs expected bank population")
ax.set_ylim(0, 1)
ax.legend(title="Income")

for p in ax.patches:
    height = p.get_height()
    ax.text(
        p.get_x() + p.get_width() / 2,
        height,
        f"{height*100:.1f}%",
        ha="center",
        va="bottom",
        fontsize=9
    )

plt.tight_layout()
plt.show()
```

```python
pi_s = pi_sample   # ~0.283 (training sample)
pi_t = pi_bank     # ~0.095 (bank)

def adjust_to_bank_prior(p, pi_s=pi_s, pi_t=pi_t):
    """
    p: model probability trained on sample (P_sample(Y=1 | x))
    returns: calibrated probability for bank (P_bank(Y=1 | x))
    """
    # Compute likelihood ratio r(x) from sample posterior
    r = (p * (1 - pi_s)) / ((1 - p) * pi_s)
    # Apply target prior
    return (pi_t * r) / (pi_t * r + (1 - pi_t))

p_grid = np.linspace(0.01, 0.99, 99)
p_corr = adjust_to_bank_prior(p_grid, pi_s, pi_t)

fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(p_grid, p_grid, "--", label="No correction (y = x)")
ax.plot(p_grid, p_corr, label="Calibrated to bank prior (9.5%)")
ax.set_xlabel("Model probability from rich sample")
ax.set_ylabel("Calibrated probability for full bank")
ax.set_title("Effect of sample bias on predicted probabilities")
ax.legend()
ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Buckets for age and hours
age_bins   = [17, 25, 35, 45, 55, 65, 90]
age_labels = ["<25", "25–35", "35–45", "45–55", "55–65", "65+"]

hour_bins   = [0, 30, 40, 45, 50, 60, 100]
hour_labels = ["<30", "30–40", "40–45", "45–50", "50–60", "60+"]

df_int["age_group"]   = pd.cut(df_int["age"], age_bins, labels=age_labels, right=False)
df_int["hours_group"] = pd.cut(df_int["hours-per-week"], hour_bins, labels=hour_labels, right=False)

marital_focus = ["Married-civ-spouse", "Never-married"]

fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)

for ax, m in zip(axes, marital_focus):
    tmp = df_int[df_int["marital-status"] == m]
    pivot = tmp.pivot_table(
        index="age_group",
        columns="hours_group",
        values="is_high_income",
        aggfunc="mean"
    ) * 100  # to %

    sns.heatmap(
        pivot,
        annot=True,
        fmt=".0f",
        cmap="Blues",
        cbar_kws={"label": "% >50K"},
        ax=ax
    )
    ax.set_title(m)
    ax.set_xlabel("Hours per week")
    ax.set_ylabel("Age group")

plt.suptitle("High-income rate by Age × Hours × Marital status", y=1.02)
plt.tight_layout()
plt.show()

segments = {
    "Married, 40–60, 45+ hrs": (
        (df_int["marital-status"] == "Married-civ-spouse") &
        (df_int["age"].between(40, 60)) &
        (df_int["hours-per-week"] >= 45)
    ),
    "Never-married, age <35": (
        (df_int["marital-status"] == "Never-married") &
        (df_int["age"] < 35)
    )
}

rows = []
for name, mask in segments.items():
    sub = df_int.loc[mask, "is_high_income"]
    rows.append({
        "segment": name,
        "n": sub.shape[0],
        "high_income_rate": sub.mean() * 100
    })

seg_df = pd.DataFrame(rows)
print(seg_df)

plt.figure(figsize=(6, 4))
sns.barplot(data=seg_df, x="segment", y="high_income_rate")
plt.ylabel("High-income rate (%)")
plt.xlabel("")
plt.title("Key segments: high-income rate by marital status + age + hours")
plt.xticks(rotation=20, ha="right")

for i, r in seg_df.iterrows():
    plt.text(i, r["high_income_rate"] + 1,
             f"{r['high_income_rate']:.1f}%\n(n={int(r['n'])})",
             ha="center")

plt.tight_layout()
plt.show()

```

```python
group_rel_sex = (
    df_int
    .groupby(["relationship", "sex"])["is_high_income"]
    .agg(["mean", "count"])
    .reset_index()
)

group_rel_sex["high_income_rate"] = group_rel_sex["mean"] * 100

# Optional: drop tiny groups if you want (e.g. Wife+Male has only 1 row)
group_rel_sex = group_rel_sex[group_rel_sex["count"] >= 50]

print(group_rel_sex.sort_values("high_income_rate", ascending=False))

plt.figure(figsize=(8, 4))
sns.barplot(
    data=group_rel_sex,
    x="relationship",
    y="high_income_rate",
    hue="sex"
)
plt.ylabel("High-income rate (%)")
plt.xlabel("Relationship")
plt.title("High-income rate by relationship × sex")
plt.xticks(rotation=20, ha="right")

plt.tight_layout()
plt.show()
seg_husband = df_int["relationship"] == "Husband"
seg_child   = df_int["relationship"] == "Own-child"

def seg_stats(mask, name):
    sub = df_int.loc[mask, "is_high_income"]
    print(f"{name}: n={sub.shape[0]}, >50K={sub.mean()*100:.1f}%")

seg_stats(seg_husband, "Husband")
seg_stats(seg_child,   "Own-child")

seg_rel = pd.DataFrame([
    {"segment": "Husband",   "rate": df_int.loc[seg_husband, "is_high_income"].mean()*100},
    {"segment": "Own-child", "rate": df_int.loc[seg_child,   "is_high_income"].mean()*100},
])

plt.figure(figsize=(4, 4))
sns.barplot(data=seg_rel, x="segment", y="rate")
plt.ylabel("High-income rate (%)")
plt.xlabel("")
plt.title("High-income rate: Husband vs Own-child")

for i, r in seg_rel.iterrows():
    plt.text(i, r["rate"] + 1, f"{r['rate']:.1f}%", ha="center")

plt.tight_layout()
plt.show()
```
